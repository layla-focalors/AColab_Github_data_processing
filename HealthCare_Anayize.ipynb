```python
# groupby( 그룹화한 결과를 dataframe으로 반환! )
df.groupby('SEX')['IDV_ID'].count()
# 가입 코드로 그룹화하여 숫자 세기 ( 가입자 코드만 )
```

```python
# 성별 키의 평균B
df.groupby('SEX')['HEIGHT'].mean()
# 성별 별 키의 평균
# 남자 키 평균이 더 크다!
```

```python
# 음주 여부에 따른 허리둘레 평균 - 전처리 x
df.groupby('DRK_YN')['WAIST'].mean()
```

```python
# 음주 여부 전처리 
df.loc[df['DRK_YN'] == 0.0, ['DRK_YN']] = 'X'
df.loc[df['DRK_YN'] == 1.0, ['DRK_YN']] = 'O'
```

```python
# 음주 여부 전처리 
df.loc[df['DRK_YN'] == 'X', ['DRK_YN']] = 'non-Drinking'
df.loc[df['DRK_YN'] == 'O', ['DRK_YN']] = 'Drinking'
```

```python
# 음주 여부에 따른 허리둘레 평균 - 전처리 x
df.groupby('DRK_YN')['WAIST'].mean()
# 평균값으로 간단하게 낸 데이터
```

```python
df.sample(20)
```

```python
df.groupby('DRK_YN')['WAIST'].describe()
# 이런 식으로 특정 열로 그룹화한 데이터에 대한 통계 총합을 낼 수 있음!
# describe() 함수는  groupby에도 적용이 가능!
```

```python
# 중요! agg 함수를 활용하여 통계를 내려고 하는 함수를 지정할 수 있음!
df.groupby('DRK_YN')['WAIST'].agg(['count', 'mean', 'median']) # 음주 여부에 따른 값 계산
```

```python
# 피벗 테이블
# 두 개 이상의 열을 행과 열의 인덱스로 설정해 데이터를 표시하는 테이블
df.pivot_table(index='SEX', aggfunc='count', values='IDV_ID')
# 인덱스는 인덱스, aggfunc= 적용할 함수, values = 어떤 데이터에 적용할지
```

```python
# 음주 여부에 따른 성별 인원수
df.pivot_table(index=['DRK_YN','SEX'], aggfunc='count', values='IDV_ID')
# 이런 식으로 그룹화하여 볼 수도 있음! ( IDVID - 가입 코드. 즉 인원수 )
```

```python
df.groupby(['DRK_YN', 'SEX'])['IDV_ID'].agg(['count','mean','median'])
# 그룹바이로도 편하게 쓸 수 있음!``
```

```python
df.pivot_table(index=['DRK_YN', 'SEX'], aggfunc=['mean','median'], values='WAIST')
# 이런 식으로 허리둘레와 음주 , 성별 별 차이를 볼 수도 있음!
```
