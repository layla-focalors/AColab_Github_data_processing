```python
# 중요! agg 함수를 활용하여 통계를 내려고 하는 함수를 지정할 수 있음!
df.groupby('DRK_YN')['WAIST'].agg(['count', 'mean', 'median']) # 음주 여부에 따른 값 계산
```

```python
# 피벗 테이블
# 두 개 이상의 열을 행과 열의 인덱스로 설정해 데이터를 표시하는 테이블
df.pivot_table(index='SEX', aggfunc='count', values='IDV_ID')
# 인덱스는 인덱스, aggfunc= 적용할 함수, values = 어떤 데이터에 적용할지
```

```python
# 음주 여부에 따른 성별 인원수
df.pivot_table(index=['DRK_YN','SEX'], aggfunc='count', values='IDV_ID')
# 이런 식으로 그룹화하여 볼 수도 있음! ( IDVID - 가입 코드. 즉 인원수 )
```

```python
df.groupby(['DRK_YN', 'SEX'])['IDV_ID'].agg(['count','mean','median'])
# 그룹바이로도 편하게 쓸 수 있음!``
```

```python
df.pivot_table(index=['DRK_YN', 'SEX'], aggfunc=['mean','median'], values='WAIST')
# 이런 식으로 허리둘레와 음주 , 성별 별 차이를 볼 수도 있음!
```

```python
# 데이터 초기로 불러오기
df = pd.read_csv('./data/heathcare/health.csv')
df.sample(10)
```

```python
df.shape
```

```python
df.columns
```

```python
# 데이터 전처리
# 필요한 컬럼 추출 : 성별, 키, 몸무게, 허리둘레, 음주 여부, 흡연 상태
# 성별, 음주 여부, 흡연 상태, 숫자 데이터를 알기 쉽게 변경
```

```python
data = df.sample(1000)
data.shape
```

```python
df = data[['SEX', 'HEIGHT', 'WEIGHT', 'WAIST', 'DRK_YN', 'SMK_STAT_TYPE_CD']]
df.shape
```

```python
df.head(10)
```

```python
df.loc[df['SEX'] == '1', ['SEX']] = 'M'
df.loc[df['SEX'] == '2', ['SEX']] = 'F'
df.loc[df['DRK_YN'] == 0, ['DRK_YN']] = 'Non-Drinking'
df.loc[df['DRK_YN'] == 1, ['DRK_YN']] = 'Drinking'
```

```python
df.loc[df['SEX'] == 1, ['SEX']] = 'M'
df.loc[df['SEX'] == 2, ['SEX']] = 'F'
df.loc[df['DRK_YN'] == 0, ['DRK_YN']] = 'Non-Drinking'
df.loc[df['DRK_YN'] == 1, ['DRK_YN']] = 'Drinking'
# 흡연 상태 전처리
df.loc[df['SMK_STAT_TYPE_CD'] == 1, ['SMK_STAT_TYPE_CD']] = 'Non-Smoking'
df.loc[df['SMK_STAT_TYPE_CD'] == 2, ['SMK_STAT_TYPE_CD']] = 'Quit-Smoking'
df.loc[df['SMK_STAT_TYPE_CD'] == 3, ['SMK_STAT_TYPE_CD']] = 'Smoking'
```

```python
df
```

```python
# seaborn 데이터 시각화
drinking = df.groupby(['SEX','DRK_YN'])['DRK_YN'].count()
drinking
```

```python
drinking = drinking.to_frame(name='COUNT')
drinking
```

```python
df_drinking = drinking.reset_index()
df_drinking
```

```python
# 성별, 흡연 상태별 인원수 계산
smoking = df.groupby(['SEX', 'SMK_STAT_TYPE_CD'])['SMK_STAT_TYPE_CD'].agg('count')
smoking
```

```python
type(smoking)
smoking = smoking.to_frame(name='COUNT').reset_index()
smoking
```

```python
type(smoking)
smoking
```

```python
print(smoking)
print(drinking)
```

```python
# 막대 그래프 그리기
ax = plt.subplot()
ax = sns.barplot(x='SEX', y='COUNT', data=drinking, errwidth=0)
ax
# errwidth - 막대기 삭제
```

```python
# 한국어 지원 설정
plt.rc('font', family='NanumGothic')
```

```python
ax = sns.barplot(x='SEX', y='COUNT', data=drinking, hue='DRK_YN')
ax.set_title('성별에 따른 음주 여부의 카운트')
```

```python
ax = sns.barplot(x='SEX', y='COUNT', hue='SMK_STAT_TYPE_CD', data=smoking)
fig = plt.figure(figsize=(10, 10))
```

```python
ax1 = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(1, 2, 2)

sns.barplot(x='SEX', y='COUNT', hue='DRK_YN', data=drinking)
sns.barplot(x='SEX', y='COUNT',  hue='SMK_STAT_TYPE_CD', data=smoking)
ax1.set_title('drinking Type')
ax2.set_title('SMKING type')
fig.suptitle('드링크와 스모킹의 타입')
fig.tight_layout()
fig
```
