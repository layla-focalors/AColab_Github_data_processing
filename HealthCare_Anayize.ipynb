```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('./data/heathcare/health.csv')
```

```python
df.info()
```

```python
subset = df[['SEX', 'WEIGHT', 'TOT_CHOLE', 'DRK_YN']]
subset
```

```python
subset.columns = ['sex', 'weight', 'tot_chole', 'drinking']
subset.loc[subset['sex'] == 1, ['sex']] = 'Male'
subset.loc[subset['sex'] == 2, ['sex']] = 'Female'
subset.loc[subset['drinking'] == 0, ['drinking']] == 'Non-Drinking'
subset.loc[subset['drinking'] == 1, ['drinking']] == 'Drinking'
subset
```

```python
subset_sample = subset.sample(1000)
subset_sample
```

```python
subset_sample.loc[subset_sample['drinking'] == 0, ['drinking']] = 'Non-Drinking'
subset_sample.loc[subset_sample['drinking'] == 1, ['drinking']] = 'Drinking'
```

```python
subset_sample
```

```python
sns.countplot(x='weight', hue='drinking', data=subset)
# hue는 범주형 차트
```

```python
# 음주가 체중에 영향을 주는 것을 알 수 있음!
# 음주 여부와 총 콜레스테롤의 관계
sns.boxplot(data=subset_sample, y='tot_chole', x='sex', hue='drinking')
```

```python
sns.violinplot(x='sex', y='tot_chole', data=subset_sample, hue='drinking')
```

```python
sns.violinplot(x='sex', y='tot_chole', data=subset_sample, hue='drinking', split=True)
```

```python
sns.swarmplot(x='sex', y='tot_chole', data=subset_sample, hue='drinking')
```

```python
sns.kdeplot(x='tot_chole', data=subset_sample)
```

```python
sns.kdeplot(subset_sample.loc[subset_sample['drinking'] == 'Drinking', 'tot_chole'], label='Drinking')
sns.kdeplot(subset_sample.loc[subset_sample['drinking'] == 'Non-Drinking', 'tot_chole'], label='Non-Drinking')
plt.legend()
# 음주 하는 경우, 콜레스테롤이 많다라는 것을 알 수 있음!
```

```python
# 체중과 총 콜레스테롤의 관계
sns.scatterplot(x='weight', y='tot_chole', data=subset_sample, hue='drinking')
```

```python
sns.lmplot(x='weight', y='tot_chole', data=subset_sample, hue='drinking')
# hue는 범주형만!
```

```python
sns.lmplot(x='weight', y='tot_chole', data=subset_sample, hue='drinking', ci=80)
# hue는 범주형만!
# 신뢰구간 : ci
```

```python
sns.lmplot(x='weight', y='tot_chole', data=subset_sample, hue='drinking', ci=None, fit_reg=False)
# hue는 범주형만!
# 신뢰구간 : ci
# lmplot 과 scatter의 차이점은 추정선의 값의 차이!
```

```python
sns.lmplot(x='weight', y='tot_chole', data=subset_sample, hue='drinking', ci=80, robust=None)
# hue는 범주형만!
# 신뢰구간 : ci
```

```python
import pandas as pd
df = pd.read_csv('./data/heathcare/health.csv')
```

```python
# 상관 분석
# 양의 상관관계 : 강(0.7 ~ 1.0) 뚜렷(0.3~0.7) 약(0.1~0.3)
# 음의 상관관계 : 강(-0.7~-1.0) 뚜렷(-0.3~-0.7) 약(-0.1~-0.3)
# 무상관 : -0.1~0.1

columns = ['AGE_GROUP', 'HEIGHT', 'WEIGHT', 'WAIST', 'TOT_CHOLE', 'HDL_CHOLE','LDL_CHOLE', 'GAMMA_GTP', 'DRK_YN']
dat = df[columns].corr()

# corr() => 상관관계 분석함수
```

```python
dat 
```

```python
dat['DRK_YN'].sort_values(ascending=False)
# sort_values와 ascending을 통해서 오름차순 정렬!
```

```python
import seaborn as sns
```

```python
sns.heatmap(dat)
```

```python
# 상관계수가 높을수록 밝은 색으로 나타남!
sns.heatmap(dat, cmap='Blues', annot=True, fmt='.2f')
# camp : 컬러맵 색상을 파랗게 지정
# annot : 상관계수 표시하기
```

```python
# 높은 상관관계를 가질수록 색이 높게 나타남!!
# 상관관계 파악 -> 히트맵!
```
